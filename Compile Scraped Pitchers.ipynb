```python
import pandas as pd
import os
```

```python
directory = os.fsencode('pitchers/')

csv_files_to_merge = []
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.startswith('pitchers'): 
        csv_files_to_merge.append('pitchers/' + filename)
        
df_list = []
for file in csv_files_to_merge:
    df_list.append(pd.read_csv(file))
    
df = pd.concat(df_list)



```

```python
def convert_height_to_inches(feet_inches):
    feet, inches = feet_inches.split('-')
    total_inches = int(inches) + (int(feet) * 12)
    return total_inches

df['HEIGHT'] = df['HEIGHT'].map(convert_height_to_inches)
df['WEIGHT'] = df['WEIGHT'].str.replace('lb','')

df = df.apply(pd.to_numeric, errors='ignore')
```

```python
df.to_csv('pitchers/all_pitching_years_1985_to_2017.csv')
```
